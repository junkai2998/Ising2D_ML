```python
### converting dat into compressed npy format

import os
import numpy as np
from random import randint
Tc=2.269

datadir="data"
pwd=os.getcwd()
temp=os.path.join(pwd, datadir)
cdir=os.listdir(temp)
print('temp=',temp)
print('cdir=',cdir)

#### abstract L #####
i=os.path.join(temp,cdir[0])
listing=os.listdir(i)
xa=[];

print(i)
print(listing)

for qi in range(len(listing)):
    if 'dat' in listing[qi]: 
        qq=listing[qi]
        xa.append(qq)

file0=xa[0];
file0=os.path.join(temp,cdir[0],xa[0])
print('The program has extracted: \n',file0)

with open(file0, "r+") as file:
    st = file.readline()
    file.close()

print ('\n the first line in the file is: \n',st)

s1=str.split(st,'; N_run')[0]
L=int(str.split(s1,"=")[1])
print("L=",L)
#### end of abstract L #####
```

```python
vstemperature=0;vsphase=(1000,1000)
x=[]

for j in os.listdir(temp):
    cdir=os.path.join(temp,j)
    print(cdir)
    listing = os.listdir(cdir)

    xa=[];
    for qi in listing:
        if 'dat' in qi: 
            xa.append(qi)
    listing=xa
    
    for i in listing:
        x.append(np.loadtxt(cdir+"/"+i))
        temperature=i.split('i')[0]
        vstemperature=np.vstack((vstemperature,temperature))
        #print(i,j,temperature)
        # putting label to each matrix with known temperature #
        temperature=float(temperature)
        #y=[];
        #y.append([1,0])
        if temperature < Tc:
            phase=[0,1]
        if temperature > Tc:
            phase=[1,0]   
        if temperature == Tc:
            phase=[randint(0,1),randint(0,1)]
#        print('temperature, phase=',temperature,phase)
        #phase=str(phase)
        vsphase=np.vstack((vsphase,phase))

# end of putting label to each matrix with known temperature #
vsphase=vsphase[1:]
vstemperature=vstemperature[1:]
x = np.array(x)
x = (x+1)/2 # convert -1 to 0, to ease ML (because ML use sigmoid, which is 0-1)
```

```python
print('shape of x_train (lattice) is: ',x.shape)
print('shape of y_train (phases) is: ',vsphase.shape)
print('shape of temperature label is:',vstemperature.shape)

#https://machinelearningmastery.com/how-to-save-a-numpy-array-to-file-for-machine-learning/
#saving array into npy compressed file
np.save("data.npy", x)
np.save("temperature.npy", vstemperature)
np.save("data_y.npy", vsphase)
#to load the file in the future simply type np.load("data.npy")
```
